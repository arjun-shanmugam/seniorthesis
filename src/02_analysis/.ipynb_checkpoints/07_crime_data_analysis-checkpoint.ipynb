```python
import pandas as pd
import geopandas as gpd
from figure_utilities import plot_labeled_hline
import dask.dataframe as dd
from dask_jobqueue import SLURMCluster
from dask.distributed import Client
import matplotlib.pyplot as plt
INPUT_DATA_CRIME = "../../data/01_raw/crime_incidents"
INPUT_DATA_TRACTS = "../../data/01_raw/Census_2010_Tracts/"
INPUT_DATA_TRACT_CHARACTERISTICS = "../../data/02_intermediate/tracts.csv"
INPUT_DATA_PANEL = "../../data/03_cleaned/crime_analysis_monthly.parquet"
```

```python
# Request computing resources.
cluster = SLURMCluster(queue='batch',
                       cores=32,
                       memory='230 GB',
                       walltime='00:20:00',
                      scheduler_options={'dashboard_address': '8787'} )
cluster.scale(jobs=1)
```

```python
client = Client(cluster)
```

```python
# Load crime data and create a column giving the month of each incident.
crime_df = crime_df = (dd.read_csv(INPUT_DATA_CRIME + "/*.csv", dtype={'REPORTING_AREA': 'object', 'SHOOTING': 'object'})
                .dropna(subset=['Long', 'Lat', 'OCCURRED_ON_DATE'])  # Drop crimes missing latitude, longitude, or date, as they cannot be merged with panel.
                .rename(columns={'OCCURRED_ON_DATE': 'month_of_crime_incident'})
                .drop(columns=['OFFENSE_CODE_GROUP', 'OFFENSE_DESCRIPTION', 'DISTRICT', 'REPORTING_AREA', 'SHOOTING', 'YEAR', 'MONTH',
                               'DAY_OF_WEEK', 'HOUR', 'UCR_PART', 'STREET', 'Location'])  # Drop unneeded columns
                .compute())
crime_df.loc[:, 'month_of_crime_incident'] = pd.to_datetime(crime_df['month_of_crime_incident'].str[:10]).dt.to_period("M").astype(str)

pre_pandemic_months = ['2019-04',
                         '2019-05',
                         '2019-06',
                         '2019-07',
                         '2019-08',
                         '2019-09',
                         '2019-10',
                         '2019-11',
                         '2019-12',
                         '2020-01',
                         '2020-02',
                         '2020-03']
crime_df = crime_df.loc[crime_df['month_of_crime_incident'].isin(['2019-01', '2019-02', '2019-03'] + pre_pandemic_months), :]
```

```python
# Plot crimes during each calendar month over the sample period.
crime_df.loc[:, 'month_of_crime_incident_without_year'] = crime_df['month_of_crime_incident'].str[-2:]

incidents_per_month = (crime_df.loc[crime_df['month_of_crime_incident'].isin(pre_pandemic_months), :]
                       .groupby('month_of_crime_incident_without_year', sort=False)
                       .count()['INCIDENT_NUMBER'])
month_display_dict = {"01": "Jan 2020", "02": "Feb", "03": "Mar", "04": "Apr 2019", "05": "May", "06": "Jun", "07": "Jul",
                      "08": "Aug", "09": "Sep", "10": "Oct", "11": "Nov", "12": "Dec"}
incidents_per_month = incidents_per_month.rename(index=month_display_dict)

fig, ax = plt.subplots()
incidents_per_month.plot(ax=ax, kind='line', color='black', xticks=range(0, 13), xlabel="Month")


mean_crimes_during_cold_months = incidents_per_month.loc[['Nov', 'Dec', 'Jan 2020', 'Feb', 'Mar', 'Apr 2019']].mean()
plot_labeled_hline(ax, y=mean_crimes_during_cold_months, text_x_location_normalized=0.35, linestyle='--', text="Mean Crime Incidents Per Month, Cold Months", color='grey', zorder=6)

mean_crimes_during_warm_months = incidents_per_month.loc[['May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']].mean()
plot_labeled_hline(ax, y=mean_crimes_during_warm_months, text_x_location_normalized=0.8, linestyle='--', text="Mean Crime Incidents Per Month, Cold Months", color='grey', zorder=6)

ax.grid(True)
plt.show()
```

```python
# Merge crimes with census tracts.
census_tracts_gdf = gpd.read_file(INPUT_DATA_TRACTS)
crime_gdf = gpd.GeoDataFrame(crime_df, geometry=gpd.points_from_xy(crime_df['Long'], crime_df['Lat'])).set_crs("EPSG:4326").to_crs("EPSG:3857")
crimes_with_tracts_gdf = gpd.sjoin(census_tracts_gdf, crime_gdf, how='inner', predicate='contains').drop(columns='geometry')
crimes_with_tracts_gdf.loc[:, 'GEOID10'] = crimes_with_tracts_gdf['GEOID10'].astype(int)
```

```python
# Merge crimes with census tract characteristics.
tract_characteristics_df = pd.read_csv(INPUT_DATA_TRACT_CHARACTERISTICS, index_col='tract_geoid')
crimes_with_tracts_gdf  = crimes_with_tracts_gdf.set_index('GEOID10')
df = crimes_with_tracts_gdf.merge(tract_characteristics_df, how='left', left_index=True, right_index=True, validate="m:1")
```

```python
poverty_rates_table = pd.DataFrame()
poverty_rates_table_index = ["All Placebo Crimes", "Shoplifting", "Motor Vehicle Accidents", "Motor Vehicle Towings", "Eviction Filings"]
poverty_rates_table_codes = [[613, 3801, 3410], [613], [3801], [3410]]
poverty_rates_table_medians = []
poverty_rates_table_75_percentiles = []
poverty_rates_table_95_percentiles = []

for code in poverty_rates_table_codes:
    median = crime_df[crime_df['OFFENSE_CODE'].isin(code)]['poor_share2010'].quantile(.5)
    poverty_rates_table_medians.append(median)
    percentile_75 = crime_df[crime_df['OFFENSE_CODE'].isin(code)]['poor_share2010'].quantile(.75)
    poverty_rates_table_75_percentiles.append(percentile_75)
    percentile_95 = crime_df[crime_df['OFFENSE_CODE'].isin(code)]['poor_share2010'].quantile(.95)
    poverty_rates_table_95_percentiles.append(percentile_95)

evictions_df = pd.read_parquet(INPUT_DATA_PANEL)
poverty_rates_table_medians.append(evictions_df['poor_share'].median())
poverty_rates_table_75_percentiles.append(evictions_df['poor_share'].quantile(.75))
poverty_rates_table_95_percentiles.append(evictions_df['poor_share'].quantile(.95))
```
